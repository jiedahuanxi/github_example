```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wiki_html = requests.get(url).text
soup = BeautifulSoup(wiki_html, 'html.parser')
# soup
```

```python
data = []
```

```python
for tr in soup.tbody.find_all("tr"):
    data.append([ td.get_text().strip() for td in tr.find_all('td')])
# data
```

```python
df = pd.DataFrame(data, columns = ['PostalCode','Borough','Neighborhood2'])
df.head()
```

```python
indexNames = df[(df['Borough'] == "Not assigned")].index
df.drop(indexNames,inplace=True)
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True)
```

```python
df=df.groupby(['PostalCode','Borough'])['Neighborhood2'].apply(', '.join).reset_index()
```

# Step 2: Use geopy library to get the latitude and longitude values of Toronto

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="trt_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

# 2.1: Create a map of Toronto

```python
# create map of Manhattan using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighborhood2']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
